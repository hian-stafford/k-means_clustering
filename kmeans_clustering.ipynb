# K-Means Clustering

```python
from numpy import random, array

def createClusteredData(N, k):
    """
    Gera dados simulados de idade e renda agrupados em clusters para um número total de pessoas distribuídas em um número de clusters.

    Args:
        N (int): Número total de pessoas.
        k (int): Número de clusters.

    Returns:
        array: Uma matriz NumPy contendo os dados gerados, onde cada linha representa uma pessoa com idade e renda.
    """
    random.seed(10)  # Define a semente para geração de números aleatórios
    pointsPerCluster = float(N)/k  # Calcula o número de pontos por cluster

    X = []  # Inicializa uma lista vazia para armazenar os dados gerados
    for i in range(k):  # Loop sobre o número de clusters
        # Gera coordenadas centrais (centroides) aleatórias para idade e renda dentro de faixas específicas
        incomeCentroid = random.uniform(20000.0, 200000.0)
        ageCentroid = random.uniform(20.0, 70.0)

        for j in range(int(pointsPerCluster)):  # Loop interno sobre o número de pontos por cluster
            # Gera valores de idade e renda aleatórios baseados nas coordenadas do centroide
            # Os valores são amostrados de uma distribuição normal com média igual ao centroide e desvio padrão específico
            X.append([random.normal(incomeCentroid, 10000.0), random.normal(ageCentroid, 2.0)])

    X = array(X)  # Converte a lista de pontos gerados em uma matriz NumPy
    return X  # Retorna a matriz de dados gerados
```

```python
%matplotlib inline

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale

# Gera dados simulados de idade e renda agrupados em clusters para 100 pessoas distribuídas em 5 clusters
data = createClusteredData(100, 5)

# Inicializa o modelo KMeans com 5 clusters
model = KMeans(n_clusters=5)

# Normaliza os dados para obter melhores resultados
data = scale(data)

# Treina o modelo KMeans nos dados
model = model.fit(data)

# Obtém os rótulos dos clusters atribuídos a cada ponto de dado
labels = model.labels_

# Imprime os rótulos dos clusters atribuídos a cada ponto de dado
print(labels)

# Visualiza os clusters
plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], data[:,1], c=labels.astype(float))
plt.show()
```
